Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.

```python
import plotly.express as px
fig = px.histogram(tip, x="sex", y="total_bill", pattern_shape = "smoker", color="day")
fig.show()
```

Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
the color parameter.

Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the
dimensions parameter.

```python
iris= sns.load_dataset("iris")
```

```python
iris
```

Q5. What is Distplot? Using Plotly express, plot a distplot.

Ans) The distplot figure factory displays a combination of statistical representations of numerical data, such as histogram, kernel density estimation or normal curve, and rug plot.

The distplot can be composed of all or any combination of the following 3 components âˆ’

1.Histogram

2.Curve: (a) kernel density estimation or (b) normal curve, and

3.Rug plot

```python
import plotly.figure_factory as ff
import numpy as np
np.random.seed(1)
x = np.random.randn(1000)
hist_data = [x]
group_labels = ['distplot']
fig = ff.create_distplot(hist_data, group_labels)
fig.show()
```
